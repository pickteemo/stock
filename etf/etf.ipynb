```python
import pandas as pd

# 指定Excel文件路径
etf_file = 'etf1118.csv'

# 使用pandas读取Excel文件
df = pd.read_csv(etf_file, dtype={'ts_code': str})
df = df.sort_values(by='3日成交额', ascending=False)
df = df.drop_duplicates(subset='细分行业', keep='first')
df['ts_code'] = df['ts_code'].apply(lambda x: x if not isinstance(x, str) else (x + '.SH' if x.startswith('5') else (x + '.SZ' if x.startswith('1') else x)))
df[['ts_code','name']].to_csv('etf.csv', index=False)
df
```

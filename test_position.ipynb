```python
import akshare as ak

hk_stock = ["00700", "09988", "01810", "02015", "09868", "09626"]
df_list = []
for code in hk_stock:
    stock_hk_hist_df = ak.stock_hk_hist(symbol=code, period="daily", start_date="20250124", end_date="20250124", adjust="")
    stock_hk_hist_df["ts_code"] = code + ".HK"
    stock_hk_hist_df = stock_hk_hist_df.rename(
            columns={
                "日期": "trade_date",
                "开盘": "open",
                "收盘": "close",
                "最高": "high",
                "最低": "low",
                "成交量": "vol",
                "成交额": "amount",
                "振幅": "amplitude",
                "涨跌幅": "pct_chg",
                "涨跌额": "change",
                "换手率": "turnover_rate",
            }
        )
    df_list.append(stock_hk_hist_df)
hk_df = pd.concat(df_list)
hk_df
```
